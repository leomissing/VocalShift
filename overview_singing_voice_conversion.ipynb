    - **Significance:**¬†*Integral part of speech synthesis, influenced by neural network approaches.*

            - *Approach:*¬†*Two-stage architecture addressing limitations of predetermined characteristics.*
        - *StarGAN v2:*
            - *Advantage:*¬†*Many-to-many conversion model without requiring large datasets.*
        - *FreeVC Algorithm:*
            - *Features:*¬†*Utilizes GANs, extends VITS method, and implements monotonic alignment search.*
- üîç¬†**Speaker Identity Extraction**
    - **Key Aspect:**¬†*D-vector usage for speaker-specific characteristics.*
    - **Enhancements:**
        - *Self-Attention Mechanism:*¬†*Used in algorithms like WavThruVec for improved speaker characteristic capture.*
        - *Generalization:*¬†*Incorporates normal distribution for better generalization.*
        - *Adversarial Networks (GANs):*¬†*Commonly applied for voice conversion.*
- üìë¬†**Linguistic Content Extraction**
    - **Focus:**¬†*Isolation of linguistic information from speech.*
    - **Strategies:**
        - *Pre-trained Models:*¬†*Utilized for linguistic feature extraction.*
        - *Variational Autoencoder (VAE):*¬†*Applied for separating linguistic characteristics.*
        - *Normalization Flow:*¬†*Implemented to improve complexity of prior distribution.*
- üéâ¬†**Generation**
    - **Critical Aspects:**
        - *Intermediate Representation:*¬†*Importance reduced in voice conversion pipeline.*
        - *Approaches:*
            - *StarGANv2-VC:*¬†*Uses GANs, style encoder, and fundamental frequency extractor.*
            - *DYGAN-VC:*¬†*Leverages dynamic convolution, WAdaIN, and VQWav2vec.*
            - *HiFi-VC:*¬†*Modifies HiFi-GAN vocoder for simplified pipeline.*
- üìà¬†**Performance Enhancements**
    - **Challenges Addressed:**
        - *Efficiency:*¬†*Smaller models for faster processing without compromising performance.*
        - *Generalization:*¬†*Improving generalization of speaker feature extractor.*
        - *Zero-shot Conversion:*¬†*Algorithms like FreeVC explore zero-shot voice conversion.*
- ü§î¬†**Current Challenges and Future Directions**
    - **Unresolved Issues:**
        - *Optimizing Models:*¬†*Balancing model size and performance.*
        - *Zero-shot VC Challenges:*¬†*Dealing with unseen target speakers without parallel training data.*
    - **Future Predictions:**
        - *Advancements:*¬†*Anticipates ongoing breakthroughs and developments.*
        - *Research Directions:*¬†*Highlights areas needing further exploration in voice conversion*
- üéô¬†**Models and Architectures:**
    - Authors of WavThruVec [11] used a GAN model with HiFi-GAN-based cost functions for linguistic and speaker information processing. D-DSVAE [14] employed a modified AutoVC model for data separation, using one encoder during training and two during conversion. WaveNet served as the vocoder. Long et al. [13] used a model similar to D-DSVAE but with mel frequency cepstral coefficients and an attention mechanism in the decoder. FreeVC [16] incorporated a posterior encoder, decoder, and discriminator using HiFi-GAN models.
- üéö¬†**Vocoders:**
    - Vocoders like Parallel WaveGAN, WaveNet, and HiFi-GAN are essential for converting speech spectrograms into sound waves.
- üåê¬†**WaveNet:**
    - WaveNet, proposed by DeepMind in 2016, is a generative model effective in natural voice sound generation, utilizing dilated convolutions and gated activation units.
- üé®¬†**Parallel WaveGAN:**
    - Parallel WaveGAN, a GAN-based model from 2020, generates realistic audio data using non-causal convolutions and a modified WaveNet.
- üöÄ¬†**HiFi-GAN:**
    - HiFi-GAN, identified as the most effective vocoder, features multi-receptive field fusion, addressing long-term dependencies and diverse periodic patterns.
- üìä¬†**Datasets:**
    - VCTK is commonly used, supplemented by datasets like Hi-Fi TTS, LibriSpeech, CommonVoice, AVSpeech, or TIMIT in various studies.
- üé§¬†**Model Inputs:**
    - Most studies use logarithmic mel spectrograms as input. Some use F0 fundamental frequency, MFCCs, or waveforms depending on the context.
- üìè¬†**Evaluation Methods:**
    - Evaluation involves objective measures (WER, CER, MCD, PCC) and subjective measures (MOS, MOSnet) for factors like linguistic consistency, prosody, and naturalness.
- üß†¬†**Challenges:**
    - Challenges include the need for extensive training data, disentangling time-variant features, handling complex aspects of human speech, and real-time performance requirements.
- üåê¬†**Conclusions:**
    - Voice conversion shows promise in various applications and continues to evolve, with the presented models demonstrating high efficiency and naturalness.
- üîç¬†**Future Directions:**
    - Future work involves selecting superior sub-solutions, addressing issues like the individual nature of human speech, real-time performance, result quality, and adaptability to different data.

# Singing Voice Conversion (SVC) Overview

But what is more important is SVC.

Here we will provide a few links for an overview:

1. (Intro) https://www.zhangxueyao.com/data/SVC/tutorial.html
2. (SOTA) https://medium.com/qosmo-lab/state-of-the-art-singing-voice-conversion-methods-12f01b35405b

**Sing voice conversion (SVC) is to convert singing voice to our desired targets**. There are three common SVC tasks:

- Intra-singer conversion: To convert a singer's voice to a desired timbre (eg: to a more beautiful voice).
- **Inter-singer conversion: To convert a singer's voice to another singers' one.**
- Cross-domain conversion: It means the source audio and target audio are from different domains. For example, to convert a speech voice to singing voice.
